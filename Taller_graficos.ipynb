```python
# Tercera celda
x=np.linspace(-10,10)
y=np.sin(x)

plt.figure()
plt.plot(x,y,'.--',label='$y=sinx$')
plt.legend()
plt.grid()
plt.show()
```

```python
# cuarta celda
x=np.linspace(-10,10,500)
y=np.sin(x**2)

plt.figure()
plt.plot(x,y,label='$y=sin x^2$',c='green')
plt.legend(loc='center right')
plt.grid()
plt.show()
```

```python
# quinta celda 

x=np.linspace(-10,10,500)
y=np.sin(x)
z=np.sin(x)**2
plt.figure()
plt.plot(x,y,'--',label='$y=sin x$',c='blue')
plt.plot(x,z,label='$y=sin^2 x$',c='green')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
# sexta celda
t=np.linspace(0,10)
x=2*t
y=2-3*t

plt.figure()
plt.plot(t,x, label='$x=x_0+vt; x_0=0, v=2$')
plt.plot(t,y,c='orange',label='$x=x_0+vt; x_0=2, v=-3$')
plt.ylabel('posicion[m]')
plt.xlabel('tiempo[s]')
plt.legend()
plt.grid()
plt.show()
```

```python
# septima celda
t=np.linspace(0,10)
x=2*t+0.5*-2*t**2
y=2-3*t+0.5*3*t**2

plt.figure()
plt.plot(t,x, label='$x=x_0+vt+0.5at^2; x_0=0, v=2, a=-2$')
plt.plot(t,y,c='orange',label='$x=x_0+vt+0.5at^2; x_0=2, v=-3,a=3$')
plt.ylabel('posicion[m]')
plt.xlabel('tiempo[s]')
plt.legend()
plt.grid()
plt.show()
```

```python
# octava celda
t=np.linspace(0,10)
x=0+2*t+0.5*-2*t**2
y=2-3*t+0.5*3*t**2

fig=plt.figure()
plt.plot(t,x, label='$x=x_0+vt+0.5at^2; x_0=0, v=2, a=-2$')
plt.plot(t,y,c='orange',label='$x=x_0+vt+0.5at^2; x_0=2, v=-3,a=3$')
plt.ylabel('posicion[m]')
plt.xlabel('tiempo[s]')
plt.legend(loc='lower left')
plt.grid()


fig.add_axes([0.25,0.6,0.2,0.2])
plt.plot(t[t<2],x[t<2])
plt.plot(t[t<2],y[t<2],c='orange')
plt.ylabel('posicion[m]')
plt.xlabel('tiempo[s]')
plt.show()
```

```python
# novena celda

x=np.linspace(10**-2,10**0)
y=2.0*x**0.5

plt.figure()
plt.loglog(x,y,label=('$T=AD^n$'))
plt.ylabel('periodo[s]')
plt.xlabel('diametro[m]')
plt.legend()
plt.grid()
plt.show()
```
